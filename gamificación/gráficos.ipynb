```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Experiencia_Sr_W.csv")
df.head(10)
```

```python
#Frecuencia de edad
plt.figure(figsize=(8, 5))
plt.hist(df['Edad'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Toma de decisiones vs. Edad
plt.figure(figsize=(10, 6))
plt.bar(df['Toma de decisiones financieras'], df['Edad'])
plt.ylabel('Edad')
plt.xlabel('Toma de decisiones financieras')
plt.title('Relación entre Toma de Decisiones y Edad')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Edad'], df['Competición'])
plt.ylabel('Competición')
plt.xlabel('Edad')
plt.title('Edad vs Competición')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Edad'], df['Gestión financiera personal'])
plt.ylabel('Gestión financiera personal')
plt.xlabel('Edad')
plt.title('Edad vs Gestión financiera')
plt.xticks(rotation=0)
plt.show()
```
